# **CONTINUOUS PROBABILITY DISTRIBUTIONS**

The relationship between the events for a continuous random variable and their probabilities is called the continuous probability distribution and is summarized by a probability density function, or PDF for short.

Unlike a discrete random variable, the probability for a given continuous random variable cannot be specified directly; instead, it is calculated as an integral (area under the curve) for a tiny interval around the specific outcome. The probability of an event equal to or less than a given value is defined by the cumulative distribution function, or CDF for short. The inverse of the CDF is called the percentage-point function and will give the discrete outcome that is less than or equal to a probability.

1. PDF: Probability Density Function, returns the probability of a given continuous outcome
2. CDF: Cumulative Distribution Function, returns the probability of a value less than or equal to a given outcome.
3. PPF: Percent-Point Function, returns a discrete value that is less than or equal to the given probability.


# NORMAL DISTRIBUTION / GAUSSIAN DISTRIBUTION

The distribution can be defined using two parameters:
1. Mean (mu or µ): The expected value.
2. Variance (sigma 2 or σ 2 ): The spread from the mean.

```python
import numpy as np
from scipy.stats import norm
from numpy.random import exponential
from scipy.stats import expon
import matplotlib.pyplot as plt
import seaborn as sns
import pylab as pl
```

```python
# generate random numbers from N(0,1)
#data_normal = norm.rvs(size=10000,loc=0,scale=1)
data_normal = np.random.normal(0, 1, 1000)
ax = sns.distplot(data_normal,
                  bins=100,
                  kde=True,
                  color='blue',
                  hist_kws={"linewidth": 20,'alpha':0.5})
ax.set(xlabel='Normal Distribution', ylabel='Frequency')

#moments of the normal distribution
mean, var, skew, kurt = norm.stats(loc=0, scale=1, moments='mvsk')
print('Mean=%.3f, Variance=%.3f, skew=%.3f, kurt=%.3f' % (mean, var, skew, kurt))
```

```python
#calculating the pmf of the distribution
x = np.linspace(norm.ppf(0.01), norm.ppf(0.99), 100)
plt.plot(x, norm.pdf(x), 'r-', lw=5, alpha=0.6, label='norm pdf')
plt.xlabel("normal variance")
plt.ylabel('pmf')
```

```python
#calculating the cdf of the distribution

x = np.arange(-4, 4, 0.01)
plt.plot(x, norm.cdf(x), 'r-', lw=5, alpha=0.6, label='norm cdf')
plt.xlabel("normal variance")
plt.ylabel('cdf in the limits (-4) to (4)')
```

```python
fig, ax = plt.subplots()
# for distribution curve
x= np.arange(-4,4,0.001)
ax.plot(x, norm.pdf(x))
ax.set_title("Cumulative normal distribution")
ax.set_xlabel('x')
ax.set_ylabel('pdf(x)')
ax.grid(True)
# for fill_between
px=np.arange(-4,1,0.01)
ax.set_ylim(0,0.5)
ax.fill_between(px,norm.pdf(px),alpha=0.5, color='g')
# for text
ax.text(-1,0.1,"cdf(x)", fontsize=20)
plt.show()
```

# EXPONENTIAL DISTRIBUTION

The exponential distribution is a continuous probability distribution where a few outcomes are the most likely with a rapid decrease in probability to all other outcomes. It is the continuous random variable equivalent to the geometric probability distribution for discrete random variables.

The distribution can be defined using one parameter:

*   **Scale (Beta or β):** The mean and standard deviation of the distribution.
*   **Rate (lambda or λ):** Rate of change in the distribution.



```python
# sample an exponential distribution
from numpy.random import exponential
from scipy.stats import expon
# define the distribution
beta = 30
n = 1000
# generate the sample
sample = exponential(beta, n)
#print(sample)

# Calculate the first four moments:
mean, var, skew, kurt = expon.stats(moments='mvsk')
print(mean, var, skew, kurt)
#histogram plotting of data distribution
exp = sns.distplot(sample,
                  bins=100,
                  kde='True',
                  color='blue',
                  hist_kws={"linewidth": 20,'alpha':1})
exp.set(xlabel='Exponential Distribution', ylabel='Frequency')
```

```python
# pdf and cdf for an exponential distribution
from scipy.stats import expon
from matplotlib import pyplot
# define distribution parameter
beta = 30
# create distribution
dist = expon(beta)
# plot pdf
values = [value for value in range(50, 70)]
probabilities = [dist.pdf(value) for value in values]
pyplot.plot(values, probabilities)
plt.xlabel('exponential distribution')
plt.ylabel('pmf')
pyplot.show()
# plot cdf
cprobs = [dist.cdf(value) for value in values]
pyplot.plot(values, cprobs)
plt.xlabel('exponential distribution')
plt.ylabel('cdf')
pyplot.show()
```

# PARETO DISTRIBUTION

A Pareto distribution is named after Vilfredo Pareto and is may be referred to as a power-law distribution. It is also related to the Pareto principle (or 80/20 rule) which is a heuristic for continuous random variables that follow a Pareto distribution, where 80% of the events are covered by 20% of the range of outcomes, e.g. most events are drawn from just 20% of the range of the continuous variable.

The distribution can be defined using one parameter:
* **Shape (alpha or α):** The steepness of the decease in probability.

Values for the shape parameter are often small, such as between 1 and 3, with the Pareto principle given when alpha is set to 1.161. We can define a distribution with a shape of 1.1 and sample random numbers from this distribution.

```python
# sample a pareto distribution
from numpy.random import pareto
# define the distribution
alpha = 1.1
n = 10
# generate the sample
sample = pareto(alpha, n)
print(sample)
```

```python
# pdf and cdf for a pareto distribution
from scipy.stats import pareto
from matplotlib import pyplot
# define distribution parameter
alpha = 1.1
# create distribution
dist = pareto(alpha)
# plot pdf
values = [value/10.0 for value in range(10, 100)]
probabilities = [dist.pdf(value) for value in values]
pyplot.plot(values, probabilities)
plt.xlabel('exponential distribution')
plt.ylabel('pmf')

pyplot.show()
# plot cdf
cprobs = [dist.cdf(value) for value in values]
pyplot.plot(values, cprobs)
plt.xlabel('exponential distribution')
plt.ylabel('cdf')

pyplot.show()
```

# UNIFORM DISTRIBUTION

The continuous uniform distribution or rectangular distribution is a family of symmetric probability distributions. The distribution describes an experiment where there is an arbitrary outcome that lies between certain bounds. The bounds are defined by the parameters, a and b, which are the minimum and maximum values. The interval can either be closed (e.g. [a, b]) or open (e.g. (a, b)) Therefore, the distribution is often abbreviated U (a, b), where U stands for uniform distribution. The difference between the bounds defines the interval length; all intervals of the same length on the distribution's support are equally probable. It is the maximum entropy probability distribution for a random variable X under no constraint other than that it is contained in the distribution's support.

```python
from scipy.stats import uniform
import matplotlib.pyplot as plt
import numpy as np

a, b = 0, 100
size = 100

# Genrating uniform distribution
uniform_distribution = uniform(loc=a, scale=b)
x = np.linspace(uniform_distribution.ppf(0), uniform_distribution.ppf(1), size)

# Plotting pdf
plt.plot(x, uniform_distribution.pdf(x), 'r-', lw=2, alpha=1, label='pdf')
plt.xlabel('uniform distribution')
plt.ylabel("pdf")
plt.show()

# Plotting cdf
plt.plot(x, uniform_distribution.cdf(x), 'r-', lw=2, alpha=1, label='pdf')
plt.xlabel('uniform distribution')
plt.ylabel("cdf")
plt.show()
```
